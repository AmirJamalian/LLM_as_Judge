```python
import re
import pandas as pd

from utilityfunctions import *
```

```python
avg_e = 1.     # Maximum avg_e which means perfect English language
avg_s = 0.2    # Minimum avg_s based on the documentation (avg_s = max(avg_s, 0.2))
```

```python
df = pd.read_csv("./essay_scores.csv")

df["q1"] = df["q1"].apply(extract_float)
df["q2"] = df["q2"].apply(extract_float)
df["q3"] = df["q3"].apply(extract_float)

df['avg_quality_scores'] = df[['q1', 'q2', 'q3']].mean(axis=1)
df['horizontal_stdevs'] = df[['q1', 'q2', 'q3']].std(axis=1)
# df['horizontal_var']    = df[['q1', 'q2', 'q3']].var(axis=1)

std_q1 = df['q1'].std()
std_q2 = df['q2'].std()
std_q3 = df['q3'].std()

avg_q = df['avg_quality_scores'].mean()
avg_h = df['horizontal_stdevs'].mean()
min_v = min(std_q1, std_q2, std_q3)

final_score = (avg_h * min_v * avg_e) / (avg_s * (9. - avg_q))

print("avg_h:", avg_h)
print("min_v:", min_v)
print("avg_e:", avg_e)
print("avg_s:", avg_s)
print("avg_q", avg_q)
print("final_score:", final_score)

display(df)
```
