```python
import pandas as pd
import numpy  as np
import os
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
topic_df_org = pd.read_csv("./topics.csv")
topic_df_org
```

```python
def generate_topic(example_topics, model_type, word_count=15):
    try:
        # Define the prompt
        prompt = (
            f"Consider these example topics, prepared for generating some essays: '{example_topics}'. "
            f"Suggest a topic (less than '{word_count}' words) like the given example but totally different from them in terms of context and area. "
        )

        # Call OpenAI API
        response = openai.ChatCompletion.create(
            model=model_type,
            messages=[
                {"role": "system", "content": "You are an expert in suggesting topics in various fields."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=word_count * 4 // 3,
            temperature=0.7                  # Adjust creativity level
        )

        # Extract and return the essay text
        topic = response['choices'][0]['message']['content'].strip()
        return topic

    except Exception as e:  # Catch generic exceptions
        return f"Error generating topic: {str(e)}"
```

```python
all_topics = topic_df_org["topic"].to_list()
all_topics
```

```python
for _ in range(5):
    new_topic = generate_topic(all_topics, "gpt-4")
    new_topic = new_topic.replace('"', '')
    all_topics.append(new_topic)
```

```python
print(len(all_topics))
all_topics
```

```python
tmp = []
for item in all_topics:
    if item[-1] == ".":
        tmp.append(item)

print(len(tmp))
tmp
```

```python
df = pd.DataFrame()
df["topic"] = tmp
df.reset_index(inplace=True)
df.rename(columns={'index': 'id'}, inplace=True)
df["id"] = df["id"].apply(lambda x: x + 1001)
df.to_csv("topics.csv", index=False)
df
```
