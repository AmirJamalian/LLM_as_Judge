```python
import random
import pandas as pd

from utilityfunctions import *
```

```python
df = pd.read_csv("./essays.csv")
```

```python
noise_sentence = "I am a blackboard. "
```

```python
for i in range(len(df)):
    rnd = random.random()
    if rnd < 1.:
        st = df.loc[i]["essay"]
        first_sentence = st.split(".")[0] + "."
        response = openai.ChatCompletion.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "You are an expert in English language."},
                {"role": "user", "content": "Negate this sentence: '{first_sentence}'"}
            ],
            max_tokens=len(first_sentence) * 4 // 3,
            temperature=0.7                  # Adjust creativity level
        )
        st = response['choices'][0]['message']['content'].strip() + " " + st
        df.at[i, "essay"] = st
        
        
        
df
```
